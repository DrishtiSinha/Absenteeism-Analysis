```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

raw_data_1 = pd.read_csv('./Absenteeism_data.csv')
raw_data = raw_data_1.copy()
raw_data
```

```python
# raw_data.info()
targets = raw_data['Absenteeism Time in Hours']
raw_data = raw_data.drop(['ID'],axis=1)
raw_data
```

```python
reasons_col = pd.get_dummies(raw_data['Reason for Absence'])
reasons_col = reasons_col.drop([0] , axis=1)

reasons_col
```

```python
raw_data = raw_data.drop(['Reason for Absence'],axis=1)
```

```python
reason_type_1 = reasons_col.loc[:,1:14].max(axis=1)
reason_type_2 = reasons_col.loc[:,15:17].max(axis=1)
reason_type_3 = reasons_col.loc[:,18:21].max(axis=1)
reason_type_4= reasons_col.loc[:,22:].max(axis=1)


data_combined = pd.concat([raw_data,reason_type_1,reason_type_2,reason_type_3,reason_type_4],axis=1)

cols = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason1', 'Reason2', 'Reason3','Reason4']
data_combined.columns = cols

reorder_cols =  ['Reason1', 'Reason2', 'Reason3','Reason4','Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']

data_combined = data_combined[reorder_cols]
data_combined
```

```python
data_combined['Date'] = pd.to_datetime(data_combined['Date'],infer_datetime_format=True)
```

```python
months = []
for i  in range(data_combined.shape[0]):
    months.append(data_combined['Date'][i].month)


data_combined['Month Values'] = months
```

```python
data_combined['Education'] = data_combined['Education'].map({1:0 ,2:1,3:1,4:1})

data_preprocess = data_combined.copy()

data_preprocess.to_csv('Absenteeism_preprocessed.csv',index=False)
```
